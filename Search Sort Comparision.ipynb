<b>Comparing time taken by linear search, binary search and bubble sort on list of 100, 200 and 1000.</b>


```python
import random
import timeit


def linear_Search(l):
    minimum=l[0]
    for i in range(len(l)):
        if l[i]<minimum:
            minimum=l[i]
    print("Minimum in the list: ",minimum)
    
def binary_Search(l,key):
    index=-1
    lower,upper=0,len(l)
    
    while lower<=upper:
        mid=(lower+upper)//2
        if key<l[mid]:
            upper=mid-1
        elif key>l[mid]:
            lower=mid+1
        else:
            index=mid
            break
    return index
        
def bubble_Sort(l):
    
    for i in range(0,len(l)):
        for j in range(0,len(l)-1-i):
            if l[j]>l[j+1]:
                temp=l[j]
                l[j]=l[j+1]
                l[j+1]=temp
    

def random_generation(n):
    l=random.sample(range(1,1000000),n)
    return l
    
def compare(num):
    l=random_generation(num)
    
    before=timeit.default_timer()
    linear_Search(l)
    after=timeit.default_timer()
    t1=after-before

    before=timeit.default_timer()
    bubble_Sort(l)
    after=timeit.default_timer()
    t2=after-before
    #print("sorting time:",after-before)
    
    # when the list size gets doubled
    l=random_generation(2*num)
    before=timeit.default_timer()
    linear_Search(l)
    after=timeit.default_timer()
    t3=after-before
    #print("Time taken in linear search:",t3)

    before=timeit.default_timer()
    bubble_Sort(l)
    after=timeit.default_timer()
    t4=after-before
    #print("sorting time:",after-before)
    
    
    l=random_generation(10*num)
    before=timeit.default_timer()
    linear_Search(l)
    after=timeit.default_timer()
    t5=after-before
    #print("Time taken in linear search:",t1)

    before=timeit.default_timer()
    bubble_Sort(l)
    after=timeit.default_timer()
    t6=after-before
    #print("sorting time:",after-before)
    print("\nLinear Search on a list of "+str(num)+" : ",t1,"seconds")
    print("Linear Search on a list of "+str(2*num)+" : ",t3,"seconds")
    print("Linear Search on a list of "+str(10*num)+" : ",t5,"seconds")
    print("\nBubble Sort on a list of "+str(num)+" : ",t2,"seconds")
    print("Bubble Sort on a list of "+str(2*num)+" : ",t4,"seconds")
    print("Bubble Sort on a list of "+str(10*num)+" : ",t6,"seconds")
    
    print("\nSEARCHING COMPARISION",t3/t1,t5/t1)
    print("SORTING COMPARISION",t4/t2,t6/t2)
    
def main():
    n=int(input("Enter the number of elements for the list: "))
    l=random_generation(n)
    before=timeit.default_timer()
    linear_Search(l)
    after=timeit.default_timer()
    print("\nTime taken in linear search:",after-before)

    before=timeit.default_timer()
    bubble_Sort(l)
    after=timeit.default_timer()
    print("\nsorting time:",after-before)
        
    k=int(input("\nEnter the key to be searched"))
    before=timeit.default_timer()
    i=binary_Search(l,k)
    after=timeit.default_timer()
    if i==-1:
        print("NOT FOUND")
    else:
        print("FOUND AT",i)
    print("searching time:",after-before)
        
    choice=int(input("\nFor comparision press 1\n "))
    if choice==1:
        compare(n)

main()
```
